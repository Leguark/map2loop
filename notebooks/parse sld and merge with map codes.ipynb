```python
import pandas as pd
import geopandas as gpd
import os

file=open('../source_data/500kibg.sld','r')
#file=open('C:/Users/00073294/Dropbox/Nishka Project/map2loop/source_data/Hammersley_Merge_plus.sld','r')
sld_file =file.readlines()
colours=[]
codes=[]
i=0
for line in sld_file:
    if('Literal>' in line):
        parts=line.split(">")
        code=parts[1].split("<")
        code=code[0]
        parts=sld_file[i+5].split(">")
        colour=parts[1].split("<")
        colour=colour[0]
        #print(code,colour)
        colours.append((colour))
        codes.append((code))
    i=i+1
    
codes_clean = [code.replace('-','_') for code in codes]

data = {'code': codes_clean, 'colour': colours,'codes2':codes}
colour_codes=pd.DataFrame.from_dict(data)
display(colour_codes)
```

```python
map=gpd.read_file('C:/Users/00073294/Dropbox/projects/map2model/full_state/500K_IBG_2016_GDB/geol_500k.shp')
#map=gpd.read_file('C:/Users/00073294/Dropbox/Nishka Project/map2loop/source_data/Hammersley_Merge_plus.shp')
```

```python
codes=map[['CODE','UNITNAME']]

map_codes=codes.drop_duplicates()

ctab=pd.merge(map_codes, colour_codes, how='inner',  left_on='CODE', right_on='codes2')
#ctab.set_index('UNITNAME',inplace=True)
ctab.drop(['CODE'], axis=1,inplace=True)
ctab.drop(['codes2'], axis=1,inplace=True)
ctab['UNITNAME'] = ctab['UNITNAME'].str.replace(' ','_')
ctab['UNITNAME'] = ctab['UNITNAME'].str.replace(',','_')
ctab['UNITNAME'] = ctab['UNITNAME'].str.replace('-','_')
display(ctab)
ctab=ctab.drop_duplicates()

ctab.to_csv('../source_data/500kibg_colours.csv',index=False)
#ctab.to_csv('C:/Users/00073294/Dropbox/Nishka Project/map2loop/source_data/Hammersley_Merge_plus_colours.csv',index=False)
```
